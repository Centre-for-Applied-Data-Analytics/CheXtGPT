```python
from PIL import Image
import numpy as np
import pandas as pd

path = '/content/image_path_resize.csv'
```

```python
def resize(image_path):
  pil_image = Image.open(image_path)
  image_width, image_height = pil_image.size
  print("Image shape:", image_width, "x", image_height)
  target_size = (299, 299)
  resized_pil_image = pil_image.resize(target_size)
  image_width, image_height = resized_pil_image.size
  print("Image shape:", image_width, "x", image_height)
  resized_pil_image.save(image_path)
```

```python
df = pd.read_csv(path)
```

```python
image_path = df['drive_path'].iloc[200]
image_path
```
